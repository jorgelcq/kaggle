```python
import geemap.foliumap as geemap
import ee
import geemap.colormaps as cmap
```

```python
ee.Authenticate()
ee.Initialize()
```

```python
Map = geemap.Map()

chile = ee.FeatureCollection("USDOS/LSIB/2017").filter(ee.Filter.eq("COUNTRY_NA", "Chile"))

landsat8 = ee.ImageCollection('LANDSAT/LC08/C02/T1_TOA')\
                                .filterDate('2022-01-01', '2023-12-31')\
                                .select(['B4', 'B5'])\
                                .filterMetadata('CLOUD_COVER_LAND', 'less_than', 10)

image = ee.Image(landsat8.median()).clip(chile)

def fun_ndvi(image):
    return image.normalizedDifference(['B5', 'B4'])

ndvi = fun_ndvi(image)

Map.addLayer(ndvi, {"palette": cmap.palettes.ndvi}, "ndvi")

Map.addLayer(ee.Image().paint(chile, 1, 2), {"palette": "red"}, "feature chile")

Map.centerObject(chile, 4)

Map
```
