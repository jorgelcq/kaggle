```r
# setwd("/kaggle/input/climatol-test")
dat <- read.csv("/kaggle/input/climatol-test/Climatol/Datos.csv")

dat <- data.matrix(dat)

head(dat)
```

```r
# Definir los datos como vectores separados
longitud <- c(-74.85, -74.62, -74.57, -74.88, -74.68, -74.12, -74.22, -73.77)

latitud <- c(-15.57, -15.40, -15.28, -14.97, -14.48, -14.68, -14.62, -15.00)

altitud <- c(19, 200, 353, 587, 1823, 3176, 3350, 3149)

identificador <- c(1, 3, 4, 5, 6, 7, 8, 9)

nombre <- c("Lomas", "Acari", "Huarato", "Copara", "Otoca", "Puquio", "Lucanas", "Coracora")

# Crear un DataFrame combinando los vectores como columnas
df <- data.frame(Longitud = longitud, 
                 Latitud = latitud, 
                 Altitud = altitud, 
                 Identificador = identificador, 
                 Nombre = nombre)

# Mostrar el DataFrame resultante
df
```

```r
write.table(df, "xyz_1964-2021.est", sep = "\t", quote = FALSE, row.names = FALSE, col.names = FALSE)

write(dat,"xyz_1964-2021.dat")
```
