```r
library(sp)
library(raster)
library(ncdf4)
library(ggplot2)
library(readr)
```

```r
# Función de extracción de data pisco de temperatura
temp_day <- function(NN, XX, YY, data_, type_date) {
    
    coor_estaciones <- data.frame(NN, XX, YY)

    # cargando archivo ncdf4
    raster_pp <- raster::brick(data_)

    # obteniendo data de archivos de entrada
    sp::coordinates(coor_estaciones) <- ~XX+YY

    # siendo
    raster::projection(coor_estaciones) <- raster::projection(raster_pp)

    # coordenadas de estaciones
    puntos_coor_estaciones <- raster::extract(raster_pp[[1]], coor_estaciones, cellnumbers = T)[,1]

    # resultado
    data_coor_estaciones <- t(raster_pp[puntos_coor_estaciones])

    # resultado
    colnames(data_coor_estaciones) <- as.character(coor_estaciones$NN)

    # Convertir la matriz en un DataFrame
    df_data <- as.data.frame(data_coor_estaciones)

    # Insertando rango de tiempo
    fechas <- seq(as.Date("1981-01-01"), as.Date("2016-12-31"), by = type_date)

    # Asignar las fechas al DataFrame
    temperatura <- as.data.frame(data_coor_estaciones, fechas)
    
    return(temperatura)
}
```

```r
# vector nombre de estaciones
NN <- c("San Geronimo")

# vector de eje de longitud
XX <- c(-75.281754)

# vector de eje de latitud
YY <- c(-11.950746)
```

```r
# temperatura diaria minima
df_min_daily <- temp_day(NN, XX, YY, "/kaggle/input/pisco-temp-min-daily/data (3).nc", "days")
head(df_min_daily)
```

```r
# temperatura diaria maxima
df_max_daily <- temp_day(NN, XX, YY, "/kaggle/input/pisco-temp-daily-nc/data (2).nc", "days")
head(df_max_daily)
```

```r
# temperatura mensual minima
df_min_monthly <- temp_day(NN, XX, YY, "/kaggle/input/pisco-temp-min-monthly/data (4).nc", "m")
head(df_min_monthly)
```

```r
# temperatura mensual maxima
df_max_monthly <- temp_day(NN, XX, YY, "/kaggle/input/pisco-temp-max-monthly/data (5).nc", "m")
head(df_max_monthly)
```

```r
write_csv(df_max_daily, "temp_max_daily.csv")
write_csv(df_min_daily, "temp_min_daily.csv")
write_csv(df_max_monthly, "temp_max_monthly.csv")
write_csv(df_min_monthly, "temp_min_monthly.csv")
```
