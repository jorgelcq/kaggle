```python
import geemap.foliumap as geemap
import ee
import geemap.colormaps as cmap
```

```python
Map = geemap.Map()

srtm4 = ee.Image("USGS/SRTMGL1_003")

dem_params = {
    "max": 550,
    "min": 0,
    "palette": cmap.palettes.viridis
}

# municipios = ee.FeatureCollection("users/jorgeluiscruzquispe7/BR_Municipios_2021")
# municipios1 = ee.FeatureCollection("users/jorgeluiscruzquispe7/BR_RG_Imediatas_2020")
# municipios2 = ee.FeatureCollection("users/jorgeluiscruzquispe7/BR_RG_Intermediarias_2020")
# estados = ee.FeatureCollection("users/jorgeluiscruzquispe7/EstadosBR_IBGE_LLWGS84")
unidadeshidrograficasbrasil = ee.FeatureCollection("users/jorgeluiscruzquispe7/UnidadesHidrograficasBrasil").\
                                    filter(ee.Filter.eq('RHI_NM', 'AMAZÔNICA'))

dem_clip = srtm4.clip(unidadeshidrograficasbrasil)


# Map.addLayer(municipios, {"color": "red"}, "municipios", opacity = 0.5, shown = False)
# Map.addLayer(municipios1, {"color": "green"}, "municipios1", opacity = 0.5, shown = False)
# Map.addLayer(municipios2, {"color": "blue"}, "municipios2", opacity = 0.5, shown = False)
# Map.addLayer(estados, {"color": "purple"}, "municipios2", opacity = 0.5, shown = False)

Map.addLayer(dem_clip, dem_params, "dem recorte")

Map.addLayer(ee.Image().paint(unidadeshidrograficasbrasil, 0, 3), {"palette": "red"}, "unidades hidrografica")


Map.centerObject(unidadeshidrograficasbrasil, 5)

Map
```
